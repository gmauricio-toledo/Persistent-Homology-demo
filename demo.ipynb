```python
import numpy as np
import matplotlib.pyplot as plt
from ripser import Rips
```

Probemos con un conjunto de puntos aleatorio

```python
data = np.random.random((100,2))

data.shape

plt.figure(dpi=120)
plt.scatter(data[:,0],data[:,1])
plt.show()
```

```python
rips = Rips()
diagrams = rips.fit_transform(data)
rips.plot(diagrams,size=23)
```

Un conjunto de puntos sobre un anillo

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=200,noise=0.15)

plt.figure(dpi=120)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
rips = Rips()
diagrams = rips.fit_transform(X)
rips.plot(diagrams,size=30)
```

Un conjunto de puntos separado en dos componentes

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=500,centers=3)

plt.figure(dpi=120)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
rips = Rips()
diagrams = rips.fit_transform(X)
rips.plot(diagrams,size=30)
```

Combinando los anteriores

```python
from sklearn.datasets import make_circles

X1, _ = make_circles(n_samples=300,noise=0.15)

noise = np.random.random(size=X1.shape)
X2 = X1 + np.array([3,3]) + 0.5*noise

noise = np.random.random(size=X1.shape)
X3 = X1 + np.array([2.5,-2]) + 0.5*noise

X = np.vstack((X1,X2,X3))

plt.figure(dpi=120)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
rips = Rips()
diagrams = rips.fit_transform(X)
rips.plot(diagrams,size=30)
```
